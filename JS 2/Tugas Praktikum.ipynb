```python
# Dependency

import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('/content/drive/MyDrive/College/ML/JS 2/Datasets/wbc.csv')
df.columns
```

```python
df_1 = df.drop(['id','Unnamed: 32'],axis=1)
df_1.head()
```

```python
le = LabelEncoder()
df_1['diagnosis'] = le.fit_transform(df['diagnosis'])
df_1.head()
```

```python
std = StandardScaler()
numerics = ['radius_mean', 'texture_mean', 'perimeter_mean',
       'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean',
       'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean',
       'radius_se', 'texture_se', 'perimeter_se', 'area_se', 'smoothness_se',
       'compactness_se', 'concavity_se', 'concave points_se', 'symmetry_se',
       'fractal_dimension_se', 'radius_worst', 'texture_worst',
       'perimeter_worst', 'area_worst', 'smoothness_worst',
       'compactness_worst', 'concavity_worst', 'concave points_worst',
       'symmetry_worst', 'fractal_dimension_worst']
df_1[numerics] = std.fit_transform(df_1[numerics])

df_1.head()
```

```python
df_1_train, df_1_left = train_test_split(df_1, test_size=0.2,random_state=0)
df_1_test, df_1_val = train_test_split(df_1_left, test_size=0.5,random_state=0)

print(f'Jumlah label data asli:\n{df_1.value_counts()}')
print(f'Jumlah label data train:\n{df_1_train.value_counts()}')
print(f'Jumlah label data unseen:\n{df_1_left.value_counts()}')
print(f'Jumlah label data val:\n{df_1_val.value_counts()}')
print(f'Jumlah label data test:\n{df_1_test.value_counts()}')
```
